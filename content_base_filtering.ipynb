```python
import plotly.express as px
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

wuzzuf_df = pd.read_csv('/content/Wuzzuf_Jobs.csv')
```

```python
developer_vacancies = wuzzuf_df.loc[wuzzuf_df['Title'].str.lower().str.contains('developer')]
print(developer_vacancies.shape)
```

```python
developer_vacancies.head()
```

```python
developer_skills = sorted(
    np.concatenate(
      developer_vacancies['Skills'].str.split(', ').tolist()
    )
)

print(list(dict.fromkeys(developer_skills)))
```

```python
skill_df = pd.DataFrame(developer_skills, columns=["Skill"]).groupby('Skill').Skill.count()
skill_df = skill_df[skill_df > 2]
skill_df = skill_df.sort_values()

fig = px.bar(skill_df, y='Skill', title="("+str(skill_df.size)+") Developer skill Occurences in vacancies")
fig.update_layout(yaxis={"dtick":10, 'range':[0,100]})
fig.show()
```

```python
#@title user skills
skills = [
    "Java 8", "Flutter",
    "ASP.NET", "Python",
    "LINQ", "Angular",
    "LARAVEL", "Laravel",
    "Typescript", "CSS", 
    "HTML", "Bootstrap",
    "JavaScript", "Javascript",
    "C#", "PostgreSQL",
    "PHP", "Dart",
    "Docker", "GIT",
]

skill_dict = dict.fromkeys(skills, 1)

print(skill_dict)
```

```python
df = developer_vacancies[['Title', 'Skills']]

for skill in skill_dict.keys():
  df.loc[df["Skills"].str.contains(skill), skill] = 1

df = df.fillna(0)
```

```python
df.head()
```

```python
vacancies_matrix = df[df.columns[2:]]
user_matrix = pd.DataFrame(skill_dict, index=[0])


df1_transposed = vacancies_matrix.T
print(user_matrix.shape)
print(df1_transposed.shape)

dot_product = user_matrix.dot(df1_transposed).T
dot_product = dot_product.sort_values(by=0, ascending=False);
dot_product = dot_product[dot_product[0] > 0]

print(dot_product.head())
```

```python
vacancies_matrix.head()
```

```python
recommended_vacancies_indexes = list(dot_product.index.values);
recommended_vacancies = developer_vacancies.loc[recommended_vacancies_indexes]
```
